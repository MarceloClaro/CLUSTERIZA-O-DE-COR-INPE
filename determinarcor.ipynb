> **k-means** - Em mineração de dados, agrupamento k-means é um método de Clustering (agrupamento ou categorização por similaridade) que objetiva particionar **n** observações dentre **k** grupos onde cada observação pertence ao grupo mais próximo da média. Isso resulta em uma divisão do espaço de dados em um **Diagrama de Voronoi**.


> - **Diagrama de Voronoi** é na matemática, um tipo especial de decomposição de um dado espaço, por exemplo, um espaço métrico, determinado pela distância para uma determinada família de objetos (subconjuntos) no espaço. Estes objetos são normalmente chamados de sítios **(cor dominante)** ou geradores (apesar de nomes como “sementes” estarem também em uso). Cada sítio está associado a **célula de Voronoi** correspondente, isto é um conjunto de todos os pontos no dado espaço o qual a distância para o dado sítio não é maior que sua distância para os outros objetos. Foi nomeado de Georgy Voronoi posteriormente, e também chamado de Tesselação de Voronoi, uma Decomposição Voronoi, ou um Mosaico de Dirichlet (em homenagem a Lejeune Dirichlet). Diagramas de Voronoi podem ser encontrados em diversos campos da ciência e tecnologia, até mesmo na arte, tendo inúmeras aplicações práticas e teóricas.

No **caso base** e nos casos mais familiares (mostrados na primeira animação 1), temos um conjunto finito de pontos ***{p1,...,pn}*** no Plano Euclidiano. Neste caso cada sítio **pk** é meramente um ponto, e corresponde a uma célula Voronoi (também chamada de região Voronoi ou célula Dirichlet) **Rk** consistindo em todos os pontos que possuem distância para pk menor do que a distância para qualquer outro sítio. Cada célula é obtida a partir da interseção de semiespaços e, portanto, é um polígono convexo. Os segmentos do diagrama de Voronoi são todos os pontos do plano equidistantes aos dois sítios mais próximos. Os vértices (nós) de Voronoi são os pontos equidistantes de três ou mais sítios.



> **Definição Formal** - Seja X um espaço (um conjunto não vazio) terminado com uma função distância d. Seja K um conjunto de índices e (Pk)k ∈ K sendo uma ***tupla*** (coleção ordenada) de subconjuntos não vazios (os sítios) no espaço X. A célula Voronoi, ou região Voronoi, Rk, associada ao sítio Pk é o conjunto de todos os pontos em X os quais a distância para Pk não é maior do que a distância para os outros sítios Pj , onde j é qualquer índice diferente de k. Em outras palavras, se d(x,A)=inf{d(x,a): a in A} denota a distância entre o ponto x e o subconjunto A, então Rk={x in X: d(x,Pk) ≤ d(x,Pj) for all j≠k}. O Diagrama de Voronoi é simplesmente a tupla de células (Rk)k ∈ K . A princípio, alguns dos sítios podem se cruzar e até coincidir (uma aplicação é descrita abaixo para os sítios que representam lojas), mas normalmente assume-se que são disjuntos. Além disso, sítios infinitos são permitidos na definição (esta definição tem aplicações na geometria dos números e cristalografia), mas, novamente, em muitos casos apenas finitamente alguns sítios são considerados.
No caso particular onde o espaço é um espaço euclidiano de dimensão finita, cada sítio é um ponto, há um número finito de pontos e todos eles são diferentes, então as células Voronoi são politopos convexos que podem ser representados de uma forma combinatória utilizando seus vértices, os lados, faces 2-dimensional, etc. Às vezes a estrutura combinatória induzida é referida como o Diagrama de Voronoi. No entanto, em geral, as células de Voronoi podem não ser convexas ou mesmo não estar conectadas.



Uma das primeiras ***aplicações do Diagrama de Voronoi*** foi por John Snow para estudar a ***epidemia de cólera*** no bairro londrino do Soho em 1854, localizado na Inglaterra. O **grupamento k-means** é um método de **Clustering**, sendo o problema computacionalmente difícil (**NP-difícil**) de decisão da soma de subconjuntos, que podemos citar o seguinte: dado um conjunto de números inteiros (cores e tons), podem algum subconjunto não-vazio formar agrupamentos de determinados tons de cores ao  pixel dominante central? Isso é uma questão do tipo sim/não, no entanto, existem algoritmos heurísticos eficientes que são comumente empregados e convergem rapidamente para um **local optimum ![local optimum](https://upload.wikimedia.org/wikipedia/commons/1/1d/Local_search_attraction_basins.png)**.



O **algoritmo KMeans** é um algoritmo típico de clustering baseado em distância, como foi dito, que usa a distância como um índice de **avaliação de similaridade**, ou seja, quanto mais próxima a distância entre dois objetos, maior a similaridade. O algoritmo acredita que os clusters são compostos de objetos próximos um do outro, portanto, **o objetivo final é obter clusters compactos e independentes**.

A **seleção de K pontos centrais do clustering** tem uma grande influência no resultado do clustering, porque na primeira etapa do algoritmo, selecione aleatoriamente quaisquer objetos k como o centro inicial de clustering, que inicialmente representa um cluster. Em cada iteração, o algoritmo atribui cada objeto restante no conjunto de dados ao cluster mais próximo, de acordo com a distância do centro de cada cluster. Quando todos os objetos de dados são examinados, uma operação iterativa é concluída e o novo centro de armazenamento em cluster é calculado.

O processo do algoritmo é o seguinte:

1. K arquivos de dados são selecionados aleatoriamente a partir de N arquivos de dados (amostras) como centróides (cluster center).
Neste artigo, os centros de cluster K são gerados aleatoriamente no espaço de amostra no processo de inicialização dos centros de cluster.
2. Meça a distância entre cada arquivo de dados e cada centróide e classifique-o no centróide mais próximo.
1.   Recalcule os centróides de cada classe que foram obtidos.
2.   Repita  (2) e (3) etapas até que o novo centróide seja igual ao centróide original ou menor que o limite especificado, o algoritmo termina.

# **Executar clustering no SciPy**
Na etapa anterior, determinamos que gostaríamos de dois clusters e agora estamos prontos para executar o cluster de k-means nos dados. Vamos criar um quadro de dados do Pandas para gerenciar facilmente as variáveis.

````python
import pandas as pd
df = pd.DataFrame({'vermelho': r,
'azul': b,
'verde': g})
````

Existem essencialmente três etapas envolvidas no processo de agrupamento de k-means com o SciPy:

Padronize as variáveis ​​dividindo cada ponto de dados pelo seu desvio padrão. Vamos usar o  whiten() método da  vq classe.
Gere centros de cluster usando o  kmeans() método
Gere rótulos de cluster para cada ponto de dados usando o  vq() método da  vqclasse
A primeira etapa acima garante que as variações em cada variável afetem os clusters igualmente. Imagine duas variáveis ​​com escalas amplamente diferentes. Se ignorarmos o primeiro passo acima, a variável com maior escala e variação teria um impacto maior na formação de clusters, tornando o processo enviesado. Portanto, padronizamos a variável usando a  whiten() função A  whiten() função pega um argumento, uma lista ou matriz dos valores de uma variável e retorna os valores padronizados. Após a padronização, imprimimos uma amostra do quadro de dados. Observe que as variações nas colunas reduziram consideravelmente nas colunas padronizadas.

Leia os dados de pixel em R, G e B e dimensione-os

````python
from scipy.cluster.vq import whiten
df['escala_vermelho'] = whiten(df['vermelho'])
df['escala_azul'] = whiten(df['azul'])
df['escala_verde'] = whiten(df['verde'])
df.sample(n = 10)
````

O próximo passo é executar o cluster de k-means com as colunas padronizadas. Usaremos a  kmeans() função de executar clustering. A  kmeans() função ( documentação ) possui dois argumentos obrigatórios - as observações e o número de clusters. Ele retorna dois valores - os centros do cluster e a distorção. Distorção é a soma das distâncias ao quadrado entre cada ponto e o centro do cluster mais próximo.

Vamos apenas usar os valores RGB dimensionados para o KMeans

````python
X = df[['escala_vermelho','escala_verde','escala_azul']].values
X
````

Como determinar o número de clusters a serem usados. Com imagens simples, é fácil identificar visualmente o número de clusters. Mas quanto maior o número de cores / gradiente na imagem, mais difícil é! O método cotovelo pode ser usado para determinar o valor ideal de k:

1: Escolha o número de clusters k;

2: Selecione k pontos aleatórios dos dados como centróides;

3: Atribua todos os pontos ao centróide mais próximo;

4: Recompute os centróides dos clusters recém-formados;

5: Repita as etapas 3 e 4;

Critérios de parada: Centroids de clusters recém-formados não mudam.

Os pontos permanecem no mesmo cluster 

Número máximo de iterações atingidas

````python
SSE = []

for cluster in range(2,8): 
    kmeans = KMeans(n_clusters=cluster,random_state=42)
    kmeans.fit(X)
    
    pred_clusters = kmeans.predict(X)
    SSE.append(kmeans.inertia_)
    
frame = pd.DataFrame({'Cluster':range(2,8) , 'SSE':SSE})
print(frame)
````

A soma dos erros quadráticos pode ser plotada para ver as alterações no SSE em relação a k. À medida que k aumenta, o cluster terá menos valores e o erro médio será reduzido. O menor número de valores em um cluster significa mais perto do centróide. O ponto em que o erro quase declina será o valor ideal para k e é o ponto do cotovelo

````python
from sklearn.datasets.samples_generator import make_blobs
from sklearn.metrics import pairwise_distances_argmin

X, y_true = make_blobs(n_samples=16000, centers= k,
                       cluster_std=0.60, random_state=0)

rng = np.random.RandomState(42)
centers = [0, k] + rng.randn(k, 2)

def draw_points(ax, c, factor=1):
    ax.scatter(X[:, 0], X[:, 1], c=c, cmap='viridis',
               s=50 * factor, alpha=0.8)
    
def draw_centers(ax, centers, factor=1, alpha=1.0):
    ax.scatter(centers[:, 0], centers[:, 1],
               c=np.arange(k), cmap='viridis', s=200 * factor,
               alpha=alpha)
    ax.scatter(centers[:, 0], centers[:, 1],
               c='black', s=50 * factor, alpha=alpha)

def make_ax(fig, gs):
    ax = fig.add_subplot(gs)
    ax.xaxis.set_major_formatter(plt.NullFormatter())
    ax.yaxis.set_major_formatter(plt.NullFormatter())
    return ax

fig = plt.figure(figsize=(15, k))
gs = plt.GridSpec(k, 15, left=0.02, right=0.98, bottom=0.05, top=0.95, wspace=0.2, hspace=0.2)
ax0 = make_ax(fig, gs[:4, :4])
ax0.text(0.98, 0.98, "BUSCANDO COR ALEATÓRIA", transform=ax0.transAxes,
         ha='right', va='top', size=10)
draw_points(ax0, 'gray', factor=2)
draw_centers(ax0, centers, factor=2)

for i in range(3):
    ax1 = make_ax(fig, gs[:2, 4 + 2 * i:6 + 2 * i])
    ax2 = make_ax(fig, gs[2:, 5 + 2 * i:7 + 2 * i])
    
    # E-step
    y_pred = pairwise_distances_argmin(X, centers)
    draw_points(ax1, y_pred)
    draw_centers(ax1, centers)
    
    # M-step
    new_centers = np.array([X[y_pred == i].mean(0) for i in range(k)])
    draw_points(ax2, y_pred)
    draw_centers(ax2, centers, alpha=0.3)
    draw_centers(ax2, new_centers)
    for i in range(k):
        ax2.annotate('', new_centers[i], centers[i],
                     arrowprops=dict(arrowstyle='->', linewidth=1))
        
    
    # Finish iteration
    centers = new_centers
    ax1.text(0.95, 0.95, "+PRÓXIMO", transform=ax1.transAxes, ha='right', va='top', size=10)
    ax2.text(0.95, 0.95, "MÉDIA", transform=ax2.transAxes, ha='right', va='top', size=10)


# Final E-step    
y_pred = pairwise_distances_argmin(X, centers)
axf = make_ax(fig, gs[:4, -4:])
draw_points(axf, y_pred, factor=2)
draw_centers(axf, centers, factor=2)
axf.text(0.98, 0.98, "AGRUPAMENTO E COR DOMINANTE DEFINIDO", transform=axf.transAxes,
         ha='right', va='top', size=10)

````

Como você pode ver, os resultados obtidos são   versões padronizadas dos valores RGB. Para obter os valores de cores originais, precisamos multiplicá-los com seus desvios padrão.

Vamos exibir as cores na forma de uma paleta usando o  imshow() método de  matplotlib's  pyplot classe. No entanto, para exibir cores,  imshow() precisa dos valores de RGB no intervalo de 0 a 1, onde 1 significa 255 em nossa escala original de valores RGB. Portanto, devemos dividir cada componente RGB de nossos centros de cluster com 255 para obter um valor entre 0 e 1 e exibi-los através do  imshow() método

Finalmente, temos mais uma consideração antes de plotar as cores usando a  imshow()função ( documentação ). As dimensões dos centros de cluster são N x 3, onde N é o número de clusters. imshow() originalmente pretende exibir uma matriz de cores AXB, portanto, espera uma matriz 3D de dimensões A x B x 3 (três elementos de cores para cada bloco na paleta). Portanto, precisamos converter nossa matriz N x 3 em matriz 1 x N x 3 passando as cores dos centros de cluster como uma lista com um único elemento. Por exemplo, se armazenamos nossas cores  colors , precisamos passar  [colors] como argumento para  imshow().

````python
import os
from collections import Counter
import colour
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
from PIL import ImageColor
def rgb2hex(rgb):
    hex = "#{:02x}{:02x}{:02x}".format(int(rgb[0]), int(rgb[1]), int(rgb[2]))
    # hex = f"#{rgb[0]}{}{}"
    return hex
rgb = [1, 2, 3]
hex = ["#"]
[hex.append(f"{c:}") for c in rgb]
''.join(hex)
PATH = './m.jpg'
WIDTH = 320
HEIGHT = 320
CLUSTERS = 5
image = Image.open(PATH)

image.size

print("Loaded {f} image. Size: {s:.2f} KB. Dimensions: ({d})".format(
    f=image.format, s=os.path.getsize(PATH)/1024, d=image.size))

image
def calculate_new_size(image):
    if image.width >= image.height:
        wpercent = (WIDTH / float(image.width))
        hsize = int((float(image.height) * float(wpercent)))
        new_width, new_height = WIDTH, hsize
    else:
        hpercent = (HEIGHT / float(image.height))
        wsize = int((float(image.width) * float(hpercent)))
        new_width, new_height = wsize, HEIGHT
        
    image.resize((new_width, new_height), Image.ANTIALIAS)
    return image, new_width, new_height
new_image, new_width, new_height = calculate_new_size(image)
print(f"New dimensions: {new_width}x{new_height}")
img_array = np.array(new_image)
img_vector = img_array.reshape((img_array.shape[0] * img_array.shape[1], 3))
new_image
model = KMeans(n_clusters=CLUSTERS)
labels = model.fit_predict(img_vector)
label_counts = Counter(labels)
model.cluster_centers_
total_count = sum(label_counts.values())
total_count
hex_colors = [rgb2hex(center) for center in model.cluster_centers_]
hex_colors
list(zip(hex_colors, list(label_counts.values())))
plt.figure(figsize=(14, 8))
plt.subplot(221)
plt.imshow(image)
plt.axis('on')
plt.subplot(222)
plt.pie(label_counts.values(),autopct='%1.1f%%', textprops=dict(size=14,color="b",weight="bold"), labels=hex_colors, colors=[color / 255 for color in model.cluster_centers_],
       shadow=True, startangle=200)
plt.axis('equal')
plt.title('ANÁLISE DE CORES DE IMAGEM SATÉLITE 02-04-19')
plt.show()
````
