## Calculando NDVI

```python
def calc_ndvi(nir,red):
    '''Calcular NDVI a partir de matrizes inteiras'''
    nir = nir.astype('f4')
    red = red.astype('f4')
    ndvi = (nir - red) / (nir + red)
    return ndvi

ndvi = calc_ndvi(nir,red)
plt.imshow(ndvi, cmap='RdYlGn')
plt.colorbar()
plt.title('Gráfico NDVI')
plt.xlabel('COLUNA #')
plt.ylabel('LINHA #')
```

### Aplicando um esquema de cores para visualizar os valores NDVI na nova imagem.
Os valores em nosso NDVI irão variar de -1 a 1. Para melhor visualizar isso, queremos usar um esquema de cores divergente e queremos centralizar a barra de cores em um ponto médio definido.

```python
from matplotlib import colors


class MidpointNormalize(colors.Normalize):
   
    def __init__(self, vmin=None, vmax=None, midpoint=None, clip=False):
        self.midpoint = midpoint
        colors.Normalize.__init__(self, vmin, vmax, clip)

    def __call__(self, value, clip=None):
       
        x, y = [self.vmin, self.midpoint, self.vmax], [0, 0.5, 1]
        return np.ma.masked_array(np.interp(value, x, y), np.isnan(value))
```

### Interpretação de NDVI

```python
# Defina os valores mín. / Máx. Do intervalo NDVI para a imagem

min= np.nanmin(ndvi)
max=np.nanmax(ndvi)

# Defina nosso ponto médio personalizado para uma análise de NDVI mais eficaz
mid=0.1

# Definindo o esquema de cores ref: https: //matplotlib.org/users/colormaps.html como referência
colormap = plt.cm.RdYlGn 
norm = MidpointNormalize(vmin=min, vmax=max, midpoint=mid)
fig = plt.figure(figsize=(40,30))
ax = fig.add_subplot(111)

# Use 'imshow' para especificar os dados de entrada, mapa de cores, mínimo, máximo e norma para a barra de cores
cbar_plot = ax.imshow(ndvi, cmap=colormap, vmin=min, vmax=max, norm=norm)

# Desligue a exibição dos rótulos dos eixos
ax.axis('off')

# Chame 'savefig' para salvar este gráfico em um arquivo de imagem
fig.savefig("/content/drive/MyDrive/Sentinel-2/NDVI-para-cluster.jpg", dpi=300, bbox_inches='tight', pad_inches=0)
fig.savefig("NDVI para cluster.jpg", dpi=300, bbox_inches='tight', pad_inches=0)

# vamos visualizar
plt.show()
```

- Primeiro, analisaremos a conversão de uma imagem em suas cores componentes na forma de uma matriz e, em seguida, executaremos o **k-means** em cluster para encontrar as cores dominantes agrupando-as.



> **k-means** - Em mineração de dados, agrupamento k-means é um método de Clustering (agrupamento ou categorização por similaridade) que objetiva particionar **n** observações dentre **k** grupos onde cada observação pertence ao grupo mais próximo da média. Isso resulta em uma divisão do espaço de dados em um **Diagrama de Voronoi**.


> - **Diagrama de Voronoi** é na matemática, um tipo especial de decomposição de um dado espaço, por exemplo, um espaço métrico, determinado pela distância para uma determinada família de objetos (subconjuntos) no espaço. Estes objetos são normalmente chamados de sítios **(cor dominante)** ou geradores (apesar de nomes como “sementes” estarem também em uso). Cada sítio está associado a **célula de Voronoi** correspondente, isto é um conjunto de todos os pontos no dado espaço o qual a distância para o dado sítio não é maior que sua distância para os outros objetos. Foi nomeado de Georgy Voronoi posteriormente, e também chamado de Tesselação de Voronoi, uma Decomposição Voronoi, ou um Mosaico de Dirichlet (em homenagem a Lejeune Dirichlet). Diagramas de Voronoi podem ser encontrados em diversos campos da ciência e tecnologia, até mesmo na arte, tendo inúmeras aplicações práticas e teóricas.


![https://upload.wikimedia.org/wikipedia/commons/0/0c/Fortunes-algorithm-slowed.gif](https://upload.wikimedia.org/wikipedia/commons/0/0c/Fortunes-algorithm-slowed.gif)

> **ANIMAÇÃO 1- DO DIAGRAMA DE VORONOI**, REPRESENTANDO A VARREDURA NA IMAGEM (TRANSFORMADA EM MATRIZ), OS PONTOS SÃO SÍTIOS - **CORES DOMINANTES**, CALCULADAS E O PROCESSO DE AGRUPAMENTO É FORMADA PELA DISTÂNCIA DOS TONS DE CORES DOMINANTES - **CENTROIDES**, ASSIM DEPOIS DE AGRUPADOS NAS **CÉLULAS DE VORONOI**, É CALCULADO A PORCENTAGEM DESSE AGRUPAMENTO EM RELAÇÃO TOTAL DE CORES.



No **caso base** e nos casos mais familiares (mostrados na primeira animação 1), temos um conjunto finito de pontos ***{p1,...,pn}*** no Plano Euclidiano. Neste caso cada sítio **pk** é meramente um ponto, e corresponde a uma célula Voronoi (também chamada de região Voronoi ou célula Dirichlet) **Rk** consistindo em todos os pontos que possuem distância para pk menor do que a distância para qualquer outro sítio. Cada célula é obtida a partir da interseção de semiespaços e, portanto, é um polígono convexo. Os segmentos do diagrama de Voronoi são todos os pontos do plano equidistantes aos dois sítios mais próximos. Os vértices (nós) de Voronoi são os pontos equidistantes de três ou mais sítios.



> **Definição Formal** - Seja X um espaço (um conjunto não vazio) terminado com uma função distância d. Seja K um conjunto de índices e (Pk)k ∈ K sendo uma ***tupla*** (coleção ordenada) de subconjuntos não vazios (os sítios) no espaço X. A célula Voronoi, ou região Voronoi, Rk, associada ao sítio Pk é o conjunto de todos os pontos em X os quais a distância para Pk não é maior do que a distância para os outros sítios Pj , onde j é qualquer índice diferente de k. Em outras palavras, se d(x,A)=inf{d(x,a): a in A} denota a distância entre o ponto x e o subconjunto A, então Rk={x in X: d(x,Pk) ≤ d(x,Pj) for all j≠k}. O Diagrama de Voronoi é simplesmente a tupla de células (Rk)k ∈ K . A princípio, alguns dos sítios podem se cruzar e até coincidir (uma aplicação é descrita abaixo para os sítios que representam lojas), mas normalmente assume-se que são disjuntos. Além disso, sítios infinitos são permitidos na definição (esta definição tem aplicações na geometria dos números e cristalografia), mas, novamente, em muitos casos apenas finitamente alguns sítios são considerados.
No caso particular onde o espaço é um espaço euclidiano de dimensão finita, cada sítio é um ponto, há um número finito de pontos e todos eles são diferentes, então as células Voronoi são politopos convexos que podem ser representados de uma forma combinatória utilizando seus vértices, os lados, faces 2-dimensional, etc. Às vezes a estrutura combinatória induzida é referida como o Diagrama de Voronoi. No entanto, em geral, as células de Voronoi podem não ser convexas ou mesmo não estar conectadas.

Uma das primeiras ***aplicações do Diagrama de Voronoi*** foi por John Snow para estudar a ***epidemia de cólera*** no bairro londrino do Soho em 1854, localizado na Inglaterra. Ele mostrou a correlação entre áreas do mapa de Londres e as áreas com mais mortes devido ao surto, utilizando uma bomba de água particular.
Uma estrutura de dados de localização de pontos pode ser construída através do Diagrama de Voronoi, já que a imagem estuda nesse projeto, possui milhares de pixels e agrupa-los por similaridade de tons e encontrar o ***centroide*** (a cor dominante) é um processo desgastante para analíse visual humana, a fim de responder a questões como o vizinho mais próximo da cor dominantes ou o objeto que está mais próximo de um ponto de determinada consulta (centroide). 
Consultas a vizinho mais próximo têm inúmeras aplicações. Por exemplo, encontrar o hospital mais próximo ou o ***objeto mais similar em um banco de dados***, que em nossa pesquisa são as cores tranformada em metrizes numéricas. Uma aplicação importante é a quantização vetorial, comumente utilizada em compressão de dados, e essencial para quantificação de cores relevantes para determinar uma cobertura vegetativa de um espaço delimitado.
Diagramas de Voronoi são utilizados também para estudar os padrões de crescimento das florestas. Também pode ser útil no desenvolvimento de modelos preditivos para incêndios florestais.

O **grupamento k-means** é um método de **Clustering**, sendo o problema computacionalmente difícil (**NP-difícil**) de decisão da soma de subconjuntos, que podemos citar o seguinte: dado um conjunto de números inteiros (cores e tons), podem algum subconjunto não-vazio formar agrupamentos de determinados tons de cores ao  pixel dominante central? Isso é uma questão do tipo sim/não, no entanto, existem algoritmos heurísticos eficientes que são comumente empregados e convergem rapidamente para um **local optimum ![local optimum](https://upload.wikimedia.org/wikipedia/commons/1/1d/Local_search_attraction_basins.png)**.



O **algoritmo KMeans** é um algoritmo típico de clustering baseado em distância, como foi dito, que usa a distância como um índice de **avaliação de similaridade**, ou seja, quanto mais próxima a distância entre dois objetos, maior a similaridade. O algoritmo acredita que os clusters são compostos de objetos próximos um do outro, portanto, **o objetivo final é obter clusters compactos e independentes**.

A **seleção de K pontos centrais do clustering** tem uma grande influência no resultado do clustering, porque na primeira etapa do algoritmo, selecione aleatoriamente quaisquer objetos k como o centro inicial de clustering, que inicialmente representa um cluster. Em cada iteração, o algoritmo atribui cada objeto restante no conjunto de dados ao cluster mais próximo, de acordo com a distância do centro de cada cluster. Quando todos os objetos de dados são examinados, uma operação iterativa é concluída e o novo centro de armazenamento em cluster é calculado.

O processo do algoritmo é o seguinte:

1. **K** arquivos de dados são selecionados aleatoriamente a partir de N arquivos de dados (amostras) como centróides (cluster center).
Neste artigo, os centros de cluster K são gerados aleatoriamente no espaço de amostra no processo de inicialização dos centros de cluster.
2. Meça a distância entre cada arquivo de dados e cada centróide e classifique-o no centróide mais próximo.
1.   Recalcule os centróides de cada classe que foram obtidos.
2.   Repita  (2) e (3) etapas até que o novo centróide seja igual ao centróide original ou menor que o limite especificado, o algoritmo termina.

# **Decodificando imagens**


As imagens podem ter várias extensões - JPG, PNG, TIFF são comuns. Esta postagem se concentra apenas nas imagens JPG, mas o processo para outros formatos de imagem não deve ser muito diferente. O primeiro passo no processo é ler a imagem.

Uma imagem com extensão JPG é armazenada na memória como uma lista de pontos, conhecidos como pixels. Um  pixel ou um elemento de imagem representa um único ponto em uma imagem. A cor do ponto é determinada por uma combinação de três valores - suas três cores componentes (vermelho, azul e verde). A cor do pixel é essencialmente uma combinação dessas três cores componentes.

Para ler uma imagem em Python, você precisa importar a  image classe de  matplotlib( documentação ). O  imread() método da classe de imagem decodifica uma imagem em seus valores RGB. A saída do  imread() método é uma matriz com as dimensões M x N x 3, onde M e N são as dimensões da imagem.

```python
img = io.imread("./NDVI para cluster.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

A matriz que obtemos do  imread() método depende do tipo de imagem que está sendo lida. Por exemplo, imagens PNG também incluem um elemento que mede o nível de transparência de um pixel. Esta postagem cobrirá apenas imagens JPG (não levanta os elementos transparência de um pixel).

Antes de passar para o agrupamento das imagens, precisamos executar uma etapa adicional. No processo de descobrir as cores dominantes de uma imagem, não estamos preocupados com a posição do pixel. Portanto, precisamos converter a matriz M x N x 3, em três listas individuais, que contêm os respectivos valores de vermelho, azul e verde. O fragmento a seguir converte a matriz armazenada na imagem em três listas individuais, cada matriz podem ter por exemplo uma área de 4715070 pixels‬ (2190 pixwls de comprimento x 2153 pixels de largura).

```python
r = []
g = []
b = []

for line in img:
    for pixel in line:
        temp_r, temp_g, temp_b = pixel
        r.append(temp_r / 255)
        g.append(temp_g / 255)
        b.append(temp_b / 255)
```

```python
df = pd.DataFrame({'vermelho': r,
'azul': b,
'verde': g})
```

O fragmento acima cria três listas vazias, e, em seguida, circula através de cada pixel da nossa imagem, acrescentando os valores RGB para a nossa  **R**,  **G** e  **B** listas, respectivamente. Se feito corretamente, cada lista terá um comprimento de 4.660.656‬ (2184 x 21340) por exemplo caso ela tenha essa dimensões.

# **Executar clustering no SciPy**
Na etapa anterior, determinamos que gostaríamos de dois clusters e agora estamos prontos para executar o cluster de k-means nos dados. Vamos criar um quadro de dados do Pandas para gerenciar facilmente as variáveis.

Existem essencialmente três etapas envolvidas no processo de agrupamento de k-means com o SciPy:

Padronize as variáveis ​​dividindo cada ponto de dados pelo seu desvio padrão. 
A primeira etapa acima garante que as variações em cada variável afetem os clusters igualmente. Imagine duas variáveis ​​com escalas amplamente diferentes. Se ignorarmos o primeiro passo acima, a variável com maior escala e variação teria um impacto maior na formação de clusters, tornando o processo enviesado. Portanto, padronizamos a variável usando a  whiten() função A  whiten() função pega um argumento, uma lista ou matriz dos valores de uma variável e retorna os valores padronizados. Após a padronização, imprimimos uma amostra do quadro de dados. Observe que as variações nas colunas reduziram consideravelmente nas colunas padronizadas.

```python
df['escala_vermelho'] = whiten(df['vermelho'])
df['escala_azul'] = whiten(df['azul'])
df['escala_verde'] = whiten(df['verde'])
# (n) é a quantidade de cores dominantes que vai ser considerada pela análise.
df.sample(n = 5)
```

```python
X = df[['escala_vermelho','escala_verde','escala_azul']].values
X
```

**GRÁFICO DE MARGENS ERROS QUADRÁTICOS / PARA ACHAR UM NÚMERO IDEAL DE CORES DOMINANTES**

A soma dos erros quadráticos (**SSE**) pode ser plotada para ver as alterações ou erros em relação a **k**. À medida que **k** aumenta, o cluster terá menos valores e o **erro médio será reduzido**. O menor número de valores em um cluster significa mais perto do centróide. O ponto em que o erro quase declina será o valor ideal para **k** e é o **ponto do cotovelo ou método cotovelo.**

```python
SSE = []

for cluster in range(2,10): 
    kmeans = KMeans(n_clusters=cluster,random_state=42)
    kmeans.fit(X)
    
    pred_clusters = kmeans.predict(X)
    SSE.append(kmeans.inertia_)
    
frame = pd.DataFrame({'Cluster':range(2,10) , 'SSE':SSE})
print(frame)
```

```python
plt.figure(figsize=(5,5))
plt.plot(frame['Cluster'],frame['SSE'],marker='o')
plt.title('Clusters x SSE')
plt.xlabel('nº de agrupamentos que melhor consegue assimilar')
plt.ylabel('(+)         <SIMILARIDADE>       (-) ')

plt.show()
```
